```python

options = webdriver.ChromeOptions()
options.binary_location = "/Applications/Google Chrome.app/Contents/MacOS/Google Chrome"
chrome_driver_binary = "./chromedriver"
browser1 = webdriver.Chrome(chrome_driver_binary, chrome_options=options)
#browser1.execute_script("alert('running javascript');")
# 終了したい場合
# browser1.quit()
```

```python
url = "https://univision.kogakkan-u.ac.jp/syllabus/"
browser1.get(url)

```

```python
handle_array = browser1.window_handles
browser1.switch_to.window (handle_array[1])
browser1.find_element(By.ID, "s_kanri_bango").find_element(By.TAG_NAME, "td").text
```

```python
for s in range(0,284):
    button = browser1.find_element_by_xpath("//*[text()='次へ']")
    button.click()
    sleep(0.08)
```

```python
colleage_name = "京都女子大学"
# ディレクトリが存在しない場合に作成
if not os.path.exists(colleage_name):
    os.mkdir(colleage_name)
handle_array = browser1.window_handles
browser1.switch_to.window (handle_array[0])

for page in range(1,286):
    target_table = browser1.find_elements(By.TAG_NAME, "table")[2]
    rows_length = len(target_table.find_elements(By.TAG_NAME, "tr"))
    print(rows_length)
    for index in range(1,rows_length):
        target_table = browser1.find_elements(By.TAG_NAME, "table")[2]
        row = target_table.find_elements(By.TAG_NAME, "tr")[index]
        pop_up_button = row.find_element(By.TAG_NAME, "a")
        class_name = row.find_elements(By.TAG_NAME, "td")[1].text + str(page).zfill(3)+str(index).zfill(3)
        print(class_name)
        if '/' in class_name:
            class_name = class_name.replace('/',"-")
        if os.path.isfile("./"+colleage_name+"/"+ class_name.replace("/", "_")+ ".html"):
            print("skip " + class_name)
        else:
            pop_up_button.click()
            sleep(1)
            handle_array = browser1.window_handles
            browser1.switch_to.window (handle_array[1])
            html = browser1.page_source
            with open("./" +colleage_name +"/"+ class_name + ".html", "a") as file:
                file.write(html)
                print(class_name)
            sleep(0.2)            
            browser1.close()
            handle_array = browser1.window_handles
            browser1.switch_to.window (handle_array[0])
            sleep(0.2)            
    button = browser1.find_element_by_xpath("//*[text()='次へ']")
    button.click()
    sleep(5)
```
