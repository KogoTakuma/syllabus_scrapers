```python
#表示されているページにある<a>タグを全て取得しそのリンクにアクセスしhtmlをダウンロードする関数
def get_html(save_path):

    sleep_time = 5

    # 保存先の作成
    def create_directory_if_not_exists(directory_path):
        if not os.path.exists(directory_path):
            os.makedirs(directory_path)
            print(f"ディレクトリ '{directory_path}' を作成しました。")
            logging.info(f"ディレクトリ '{directory_path}' を作成しました。")   
        else:
            print(f"ディレクトリ '{directory_path}' は既に存在します。")

    directory_path = "./html" + save_path 
    create_directory_if_not_exists(directory_path)

    

    # <a>要素を取得します
    a_elements = browser1.find_elements(By.TAG_NAME, 'a')



    # 取得した<a>要素のテキストを出力します
    for element in a_elements:
        try:
            raw_code = element.get_attribute("onClick")
            start_index = raw_code.find("'") + 1
            end_index = raw_code.find("'", start_index)
            substring = raw_code[start_index:end_index]
            key = re.sub(r'^../../../..', "", substring)
            url = "https://syllabus3.jm.kansai-u.ac.jp/syllabus/search" + key

            response = requests.get(url)
            html = response.text
            #time.sleep(sleep_time)
            # htmlをファイルに保存
            file_name = key.replace('/', '-')
            with open(r""+directory_path + file_name, 'w', encoding='utf-8') as f:
                f.write(html)
                print(file_name)
                logging.info("complete downloading" +file_name)
        except AttributeError as e:
            pass
```

```python
def select_option(browser, dropdown_name_list, selected_options):

    #選択肢の取得
    dropdown = browser.find_element(By.ID, dropdown_name_list[0])
    select = Select(dropdown)
    option_list = []
    for option in select.options:
        option_list.append(option.text)
    #選択肢の入力
    for option in option_list:
        dropdown = browser.find_element(By.ID, dropdown_name_list[0])
        select = Select(dropdown)
        select.select_by_visible_text(option)
        time.sleep(0.2)
        this_selected_options = selected_options.copy()
        this_selected_options.append(option)

        if len(dropdown_name_list) == 1:
            try:
                time.sleep(2)
                search_button = browser1.find_element(By.NAME, "submitSearch")
                search_button.click()
                title = ""
                for selected_option in this_selected_options:
                    title = title + selected_option + "/"
                title = title.replace("----------/","")

                time.sleep(2)
                get_html(title)
                print(title)

                "come"
                back_button = browser1.find_element(By.NAME, "submit2")
                back_button.click()
            
            except UnexpectedAlertPresentException as e:
                # UnexpectedAlertPresentExceptionが発生した場合の処理を記述する
                print("UnexpectedAlertPresentExceptionが発生しました:")
                sleep(1)
        else:
            next_dropdown_name_list = dropdown_name_list.copy()
            next_dropdown_name_list.pop(0)
            
            select_option(browser, next_dropdown_name_list, this_selected_options)


```
